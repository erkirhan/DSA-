# 3. Who scores the most on average for math, reading and writing based on

# 4. What do you think about the scoring variation for math, reading and writing based on

```python
fem=data[data['gender']=='female']
mal=data[data['gender']=='male']
```

```python
fem.plot(kind='line', y='math score')
plt.title("Female math scores")
fem.plot(kind='line', y='reading score')
plt.title("Female reading scores")
fem.plot(kind='line', y='writing score')
plt.title("Female writing scores")
```

```python
tst1=data[data['test preparation course']=='none']
tst2=data[data['test preparation course']=='completed']
```

```python
tst1.plot(kind='line', y='math score')
plt.title("none math scores")
tst1.plot(kind='line', y='reading score')
plt.title("none reading scores")
tst1.plot(kind='line', y='writing score')
plt.title("none writing scores")
```

```python
tst2.plot(kind='line', y='math score')
plt.title("completed math scores")
tst2.plot(kind='line', y='reading score')
plt.title("completed reading scores")
tst2.plot(kind='line', y='writing score')
plt.title("completed writing scores")
```
